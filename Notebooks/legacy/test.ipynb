```python
def find_stable_region(array, window_size):
    if len(array) < window_size:
        raise ValueError("Window size must be smaller than or equal to the length of the array")

    variances = []
    for i in range(len(array) - window_size + 1):
        window = array[i:i + window_size]
        variance = np.var(window)
        variances.append(variance)
    
    min_variance_index = np.argmin(variances)
    stable_region = array[min_variance_index:min_variance_index + window_size]
    median_value = np.median(stable_region)
    
    return stable_region, min_variance_index, min(variances), median_value
```

```python
# Empirical survival function
sorted_data = np.sort(data)
empirical_survival = 1 - np.arange(1, len(data) + 1) / len(data)

# Theoretical survival function using the fitted Weibull parameters
theoretical_survival = 1 - stats.weibull_min.cdf(sorted_data, shape, loc, scale)

# QQ Plot
plt.figure(figsize=(8, 6))
plt.loglog(theoretical_survival, empirical_survival, marker='x', linestyle='', markersize=8)
plt.plot([min(theoretical_survival), max(theoretical_survival)], 
         [min(theoretical_survival), max(theoretical_survival)], 'r--', label='Diagonal')
plt.xlabel('GPD survival function')
plt.ylabel('Empirical survival function')
# make the x and y axis descending 
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.title('Q-Q plot against fitted Weibull distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from thesis_tools.utils.data import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
region = "Central Eurasia"
year = "2022"
df_selfmade = read_billionaires_data(
    only_years=[year],
    only_regions=[region],
    self_made=True
)
df_inherited = read_billionaires_data(
    only_years=[year],
    only_regions=[region],
    self_made=False
)
```

```python
df_selfmade
```

```python
# histogram
fig, ax = plt.subplots()
n_bins = 100
ax.hist(df_selfmade["net_worth"], bins=n_bins, alpha=0.5, label="Self-made")
ax.hist(df_inherited["net_worth"], bins=n_bins, alpha=0.5, label="Inherited")
ax.set_xlabel("Net worth")
ax.set_ylabel("Frequency")
# fit density
ax.legend()
plt.show()
```

```python
# Fit density function
# Overlay KDE plot
fig, ax = plt.subplots()
sns.kdeplot(df_selfmade["net_worth"], label="Self-made", ax=ax)
sns.kdeplot(df_inherited["net_worth"], label="Inherited", ax=ax)
ax.set_xlabel("Net worth")
ax.set_ylabel("Density")
ax.legend()
plt.show()
```

```python
# histogram
fig, ax = plt.subplots()
n_bins = 100
ax.hist(df_selfmade["log_net_worth"], bins=n_bins, alpha=0.5, label="Self-made")
ax.hist(df_inherited["log_net_worth"], bins=n_bins, alpha=0.5, label="Inherited")
ax.set_xlabel("Net worth")
ax.set_ylabel("Frequency")
# fit density
ax.legend()
plt.show()
```
